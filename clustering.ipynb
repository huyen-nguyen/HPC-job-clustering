#### With non-scaled

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances

features = ['N_n', 'M_u', 'T_q', 'T_r']

df = pd.read_csv('data/data copy.csv', skipinitialspace=True, usecols=features)

kmeans = KMeans(n_clusters=4, precompute_distances="auto", n_jobs=-1)

label = kmeans.fit_predict(df[features])

df['Cluster'] = label

print(label)

# df.to_csv(r'result/kmeans-3.csv', index = False)

# ====== PCA ======
reduced_data = PCA(n_components=2).fit_transform(df[features])
results = pd.DataFrame(reduced_data,columns=['pca1','pca2'])

sns.scatterplot(x="pca1", y="pca2", hue=label, data=results)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```
